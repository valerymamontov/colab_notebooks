### Поиск битых ссылок на сайте МИАЦ Калужской области  

Задача - найти все битые ссылки на сайте.

```
# подключение необходимых библиотек

from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```
# создаются два "набора", т.е. два множества, которые будут содержать уникальные значения:
# в 1-ом множестве - все найденные страницы (ссылки), без повторов
# во 2-ом множестве - все битые ссылки, без повторов

pages = set()
badlinks = set()



# функция, которая в качестве параметра принимает стартовую страницу
# затем собирает все ссылки на странице и рекурсивно передаёт страницу в функцию


def getLinks(pageUrl):
    html = urlopen(r"http://miac.kaluga.ru/"+pageUrl)
    soup = BeautifulSoup(html)

    for a in soup.findAll("a", href=re.compile("^/")):
        if "href" in a.attrs and a.attrs["href"] not in pages:
            link = a.attrs['href']
               
            try:
                urlopen(f"http://miac.kaluga.ru{link}")
            except:
                if link not in badlinks:
                    badlinks.add(link)
                print(f"bad link: {link}")
                continue
            else:
                pages.add(link)
                print(link)
                getLinks(link)
```
